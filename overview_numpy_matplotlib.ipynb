### Numpy y matplotlib

```python
#Ejemplo 1
import matplotlib.pyplot as plt #submodulo de matplotlib para graficar

x = [0.2,0.8,1.3,2.5,2.5]
y = [0.5,0.6,0.8,0.9,0.9]

plt.plot(x,y) #grafica de puntos
plt.title("Grafica de puntos")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show() #muestra la grafica  
```

```python
#Ejemplo 2
vec2 = np.linspace(-5,5,100) #crea un vector de 100 puntos entre -5 y 5
type(vec2)
#Numpy realiza todas las operaciones de manera vectorizada
```

```python
vec4 = vec2 +0.5 #suma 0.5 a cada elemento del vector
print(vec4)
```

```python
#pasar a radianes 
vec3 = np.radians(vec2) #convierte los grados a radianes
vec3 = np.sin(vec3) #aplica la funcion seno a cada elemento del vector
print(vec3)
```

```python
#acceder a sus elementos mediante indices
print(vec3[0:10]) #muestra los primeros 10 elementos del vector
```

```python
#valor maximo y minimo
print("Valor maximo:", vec3.max())
print("Valor minimo:", vec3.min())
```

Conjunto de partida (dominio):
$$
x \in [-3,3]
$$
$$
y = \sin(x^{2}) \cdot \exp(x^{2})
$$

```python
#definamos la variable independiente x
x = np.linspace(-3,3,100)

#definamos la variable dependiente y
y = np.sin(x**2) * np.exp(x**2)
#diagrama de dispersión
plt.plot(x,y)
plt.show()
```

```python
#consideremos el definir los valores de la variable dependiente utilizando una funcion
def f1(x: np.ndarray) -> np.ndarray:
    return np.sin(x**2) * np.exp(x ** 0.5)

#definamos la variable independiente x
x1 = np.linspace(2,8,600)
#definamos la variable dependiente y
y1 = f1(x1)
#diagrama de dispersión
plt.plot(x1,y1)
plt.show()
    
```

```python
#Ejemplo 2: dos reglas de correspondencia en una misma ventana (grafica)
#definamos la variable independiente x
x2 = np.linspace(2,18,120)

#definamos las dos reglas de correspondencia
y2_a = np.sin(x2**(1/7))
y2_b = np.exp(-x2 ** 2)

#en el mismo grafico colocamos las dos reglas de correspondencia
plt.plot(x2, y2_a, x2, y2_b)
plt.title("Dos reglas de correspondencia")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend(["Regla de correspondencia A", "Regla de correspondencia B"])
plt.show() #muestra la grafica
```

```python
#Ejemplo 3: 

#definamos un conjunto de partida
x3 = np.linspace(np.pi/3, 5, num = 600)

#definamos una regla de correspondencia para representar una señal limpia
y3 = np.exp(np.sqrt(x3**(1/8)))

#a la señal limpia le agregamos ruido
y3_ruido = y3 + np.random.normal(0, 0.01, len(x3))

#mostremos el diagrama de dispersion
plt.plot(x3, y3, "yellow", x3, y3_ruido, "blue")

plt.show()
```

```python
#Ejemplo 4: mostrar grafica limpia y sin ruido por separado
#ventana 1
plt.figure(num=1)
plt.plot(x3, y3, "yellow")
plt.title("Señal limpia")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
#ventana 2
plt.figure(num=2)
plt.plot(x3, y3_ruido, "blue") 
plt.title("Señal con ruido")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```
