## Creacion de datos de tipo Series


```python
# Construyamos un objeto unidimensional de pandas a partir de un ndarray de numpy
v = np.array([1,3,4,6,2,3,0])
s1 = pd.Series(v)
s1
```

```python
# Lista de metodo que se puede aplicar a un dato de tipo pd.Series
dir(s1)
```

```python
# Observacion : la funcion dir devuelve metodos y atributos
```

```python
# Los atributos no requieren parentesis
help(s1.dtype)
```

```python
# Los metodos si requieren parentesis : Indica la posibilidad de requerir argumentos
help(s1.isnull)
```

```python
# datatype de los elementos que compone al Series
s1.dtype
# Todos los elementos que componen a s1 son de tipo int64
```

```python
s1.itemsize
```

```python
# numero de bytes consumido en memoria ram por el objeto s1
s1.nbytes
```

```python
# En vista de los explicado y por una caracteristica inherentemente hereditaria concluimos
# que podemos utilizar la funcion len para el calculo del numero de elementos de un dato de
# tipo Series
len(s1)
```

```python
# El atributo shape
s1.shape
```

```python
# El metodo count
s1.count()
```

```python
# El atributo size
s1.size
```

```python
# numero de dimensiones
s1.ndim
```

```python
# Tambien podemos crear un dato de Series usando un dict
d1 = {"a1":2, "a2":5,"a3":8,"a4":10}
s2 = pd.Series(d1)
s2
```

```python
s3 = pd.Series([12,11,9,14,0],index=["n1", "n2","n3","n4", "n5"])
s3
```

```python
s4 = pd.Series(np.random.randint(0,20,5),index=["n1", "n2","n3","n4", "n5"])
s4
```

```python
# Creacion de un DataFrame a partir de una lista de diccionario
data = [{"a":12, "b":15}, {"c":2, "b":9,"a":18}]
df2 = pd.DataFrame(data)
df2
```

```python
# MOdificamos los nombres de las columnas
df2.columns = ["eval1", "eval2","eval3"]
df2
```

```python
# MOdificamos las etiquetas que tienen las filas
df2.index = ["st1", "st2"]
df2
```

```python
# Definamos las etiquetas para filas y columnas de un dataframe al momento de creacion
df3 = pd.DataFrame(data, index = ["N1","N2"], columns = ["O1", "O2", "O3"])
df3
```

```python
# Definamos las etiquetas para filas y columnas de un dataframe al momento de creacion
df4 = pd.DataFrame(data, index = ["N1","N2"], columns = data[1].keys())
df4
```

```python
# Calculemos el minimo de la columna price
df.price.min()
```

```python
# Calculemos el maximo de la columna price
df.price.max()
```

```python
# Calculemos alguno de los percentiles
df.price.quantile(0.75)
```

```python
# Podemos filtrar informacion a partir de un percentil
# Obtengamos la informacion que cumple la condicion : price < perciltil 75 de price
df[df.price > df.price.quantile(0.75)]
```

```python
# Ya aprendimos que hay un precio maximo de la columa price, obtengamos la informacion
# de las observaciones que cumplen esa condicion
df[df.price == df.price.max()]
```

```python
# Analicemos a la columna year
# minimo de los valores no NaN
df.year.min(), df.year.max()
```

```python
# Filtermos la informacion del 2020 en adelante
df[df.year >= 2020]
```

```python
# Precio promedio de los carros publicados despues del 2020
df[df.year >= 2020].price.mean()
```
